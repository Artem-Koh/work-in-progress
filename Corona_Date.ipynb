```python
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
import urllib3
import os
import datetime
```

```python


h = urllib3.PoolManager()
resp = h.request('GET', 'https://gogov.ru/covid-19/msk')
a = resp.data.decode('utf-8')
soup = BeautifulSoup(a, 'lxml')
date = []
quantity = []
new_cases = []
j = 0
y = 0
for tag in soup.find_all("tr"):
        x = str("{0}: {1}".format(tag.name, tag.text).strip('tr: '))
        
        if len(x) > 40:
            date += [x[:8]]
            x = x[8:]
            x = re.findall(r'[^а-я]*', x)[0]
            if len(x) > 1:
                while x[j] != '(':
                    j += 1
                quantity += [x[:j]]
                x = x[j:]
                while x[y] != ')':
                    y += 1
                new_cases += [x[1:y]]
            y = 0
            j = 0
        else:
            break
date = date[1:32]
quantity = quantity[:31]
new_cases = new_cases[:31]
```

```python
table = pd.DataFrame([quantity, new_cases], 
columns = [date])

table
```
