```python
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
import urllib3
```

```python
h = urllib3.PoolManager()
resp = h.request('GET', 'https://gogov.ru/covid-19/msk')
a = resp.data.decode('utf-8')
soup = BeautifulSoup(a, 'lxml')
date = []
quantity = []
new_cases = []
j = 0
y = 0
for tag in soup.find_all("tr"):
        x = str("{0}: {1}".format(tag.name, tag.text).strip('tr: '))
        x = re.findall(r'[^а-я]*', x)[0]
        if len(x) > 40:
            date += [x[:8]]
            x = x[8:]
            quantity += [x[:7]]
            x = x[7:]
            new_cases += [x[2:8]]
            y = 0
            j = 0
date = date[:31]
quantity = quantity[:31]
new_cases = new_cases[:31]
```

```python
table = pd.DataFrame([quantity, new_cases], 
columns = [date])

table
```
